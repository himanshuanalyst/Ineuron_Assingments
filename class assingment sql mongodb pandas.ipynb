```python
 # 7. Write a sql query to find out how many unique dress that we have based on dress id
```

```python
import mysql.connector as connection
mydb = connection.connect(host="localhost",user="root", passwd="ramram",database="fsds_data",use_pure=True)
```

```python
query = '''
 SELECT count(distinct(dress_sales.Dress_ID))
 FROM dress_attribute LEFT JOIN dress_sales
 ON dress_attribute.Dress_ID=dress_sales.Dress_ID
'''
cursor = mydb.cursor() #create a cursor to execute queries
cursor.execute(query)
print(cursor.fetchall())
```

```python
#  9. Try to find out total dress sell for individual dress id
```

```python
query = '''
 select Dress_ID,sum(`29/8/2013`+`31/8/2013`+`09-02-2013`+`09-04-2013`+`09-06-2013`+`09-08-2013`+ `09-10-2013`+ `09-12-2013`+
  `14/9/2013`+ `16/9/2013`+ `18/9/2013`+ `20/9/2013`+ `22/9/2013`+ `24/9/2013` + `26/9/2013`+ `28/9/2013`+ `30/9/2013`+ 
  `10-02-2013`+`10-04-2013`+`10-06-2013`+`10-08-2010`+`10-10-2013`+`10-12-2013`) Total_sales from dress_sales
group by Dress_ID
order by Total_sales desc;
'''

pd.read_sql_query(query,engine)
```
